### Feature Extraction-Classification-Relu Activation- Pooling Size reducing

ANN not used for image classification if too much pixels(big image) huge number of weights to be found... Sensitive to image location like can detect same face in center not side

CNN: Different set of neurons predict specific features and club to predict head middle and tail...Independent of Image location

Working: Take specific size(3 by 3) grid from original image and multiply with filter(3 by 3) and store average in feature map....... Stride=1 Every next grid

Feature map=1 or closest to one shows it passes the filter criteria

Max Pooling (4 by 4) size to (2 by 2) half size by selecting max value in half grid size stride 2... Pooling reduces computation size, overfitting less parameters, tolerant to variation filters out noise

Disadvantage: Rotation and Scale(Thickness) not taken care of... Use Data Augmentation or train model with such samples also
